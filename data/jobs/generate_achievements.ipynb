```python
import os
from groq import Groq
import dotenv

dotenv.load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API_KEY"),
)
```

```python
experience_levels = {}

for row in df.itertuples():
    if not experience_levels.get(row.formatted_experience_level):
        experience_levels[row.formatted_experience_level] = []
    experience_levels[row.formatted_experience_level].append(row.title)
```

```python
kept_jobs = []

for row in df.itertuples():
    if row.formatted_experience_level in ['Associate', 'Internship', 'Entry level'] or type(row.formatted_experience_level) != str:
        kept_jobs.append(row)
        
print(set([job.company_name for job in kept_jobs]))
```

```python
def generate_prompt(job):
    return f"""job description: {job.description}
company name: {job.company_name}
job title: {job.title}

I am building a fake CS student resume generator. The above data is a job posting. Generate a list of plausible achievements that a college student working in this role could include on their resume. Include ONLY achievements the college student would have done on the job, not just any achievement. Each achievement should be concise, specific, and realistic for someone with limited professional experience. Put each achievement on a separate line with no extra formatting. Include no text besides the achievements.

If the job description is too advanced for a current college student or a recent graduate (that is, if the job is too senior), return the word "ADVANCED" in all caps with no additional text.
Include no text besides the generated achievements or the word "ADVANCED"
"""
```

```python
generate_prompt(kept_jobs[0])
```

```python
kept_jobs[0]
```

```python
def generate_achievements(job):
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": generate_prompt(job),
            }
        ],
        model="llama3-8b-8192",
        stream=False,
    )

    return chat_completion.choices[0].message.content
```

```python
generate_achievements(kept_jobs[5])
```

```python
import time
from tqdm import tqdm  

achievements_dict = {}

for job in tqdm(kept_jobs, desc="Generating achievements", unit="job"):
    achievements_dict[job.job_id] = generate_achievements(job)
    time.sleep(2) 
```

```python
final_adict = {}

for k, v in achievements_dict.items():
    if "ADVANCED" in "\n".join(v):
        continue
    new_v = []
    for achv in v:
        if not achv: # Empty string
            continue
        
        # Remove bullet points
        achv = achv.lstrip("*•- ").strip()
        
        if achv:
            new_v.append(achv)
    final_adict[k] = new_v
```

```python
import json

with open("company_main_pages.json", "r") as f:
    sites = json.load(f)

SENIOR_STUFF = ['senior', 'sr', 'principal']

def too_senior(job):
    return any(word in job.lower() for word in SENIOR_STUFF)

final_jobs_json = {}

for job in kept_jobs:
    if job.job_id in final_adict and not too_senior(job.title):
        final_jobs_json[str(job.job_id)] = {
            "company_name": job.company_name,
            "job_title": job.title,
            "achievements": final_adict[job.job_id],
            "location": job.location,
            "site": sites[job.company_name]
        }
        
with open("final_jobs_info.json", "w") as f:
    json.dump(final_jobs_json, f, indent=4)
```
